# **SINGLE SAMPLE T-TEST USING TITANIC DATASET**

**MY ASSUMPTION IS THE POPULATION MEAN AGE IS 35**

```python

import numpy as np 
import pandas as pd 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train = pd.read_csv('/kaggle/input/titanic/train.csv').drop(columns = ['Survived'])
test =  pd.read_csv('/kaggle/input/titanic/test.csv')
```

```python
final_df = pd.concat([train , test]).sample(1309)
```

```python
pop = final_df['Age'].dropna()
```

```python
pop.mean()
```

**CHOOSING SAMPLE**

```python
sample_age = pop.sample(25).values
```

```python
sample_age
```

**NULL HYPOTHESIS**:  H0 -> The mean age is 35

**ALTERNATIVE HYPOTHESI**S:  H1 -> The mean is less than 35

Now since the sample size is less than 30 so we will use **Shapiro Wilk test** to check normality of given distribution

```python
from scipy.stats import shapiro

shapiro_age = shapiro(sample_age)

print(shapiro_age)
```

```python
pop_mean = 35
```

**PERFORMING THE TEST**

```python
import scipy.stats as stats

t_statistic, p_value = stats.ttest_1samp(sample_age, pop_mean)

print("t-statistic:", t_statistic)
print("p-value:", p_value/2)
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```
